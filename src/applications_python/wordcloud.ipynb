```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

with open("../../exos/nee.txt") as f:
    text = f.read()
for s in '.', '!', ',', '?', ':', '[', ']', 'ARTHUR', 'HEAD KNIGHT', 'PARTY':
    text = text.replace(s, '')
text
```

```python
# Create the wordcloud object
wcloud = WordCloud(width=480, height=480, margin=0).generate(text)
 
# Display the generated image:
plt.imshow(wcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
image = wordcloud.to_image()
image.show()
```

```python
#https://si.wsj.net/public/resources/images/HC-GW391_Drawin_G_20180828103522.jpg
# read the mask / color image taken from
# http://jirkavinse.deviantart.com/art/quot-Real-Life-quot-Alice-282261010
from PIL import Image
import numpy as np
from wordcloud import WordCloud, STOPWORDS
python_mask = np.array(Image.open("python_source.png"))
stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="white", max_words=2000, mask=python_mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue')

# generate word cloud
wc.generate(text)

# store to file
wc.to_file("python.png")

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(python_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
